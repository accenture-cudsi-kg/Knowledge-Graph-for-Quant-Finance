```python

from nltk.tokenize import sent_tokenize

def baseline_findoc(documents):
    final_output = {}
    for doc_key in documents:
        output = []
        text = documents[doc_key]

        # Perform coreference resolution
        text_coref = coref.coref_resolved(document=text)

        # Break the paragraph into sentences
        sent_coref = sent_tokenize(text_coref)

        # Make a sentence-by-sentence prediction
        for sent in sent_coref:
            srloutput = srlbert.predict(sent)
            output.extend(list(map(lambda x: x['description'], srloutput['verbs'])))
        print(f'doc: {doc_key}, before filtering: {len(output)}')

        # Filter on outputs that include 'ARG'
        filtered = list(filter(lambda x: '[ARG' in x, output))
        print(f'doc: {doc_key}, after filtering: {len(filtered)}')
        final_output[doc_key] = filtered
    return final_output
```

```python
final_output = baseline_findoc(fin_documents)
```
